```python
import json
import pandas as pd
with open("./pitches.json") as f:
    j = json.load(f)


def prepare(arr):
    for page in arr:
        try:
            yield {"src_url": page["src_url"], "url": page["url"]["href"], **page["initialState"]["opportunity"]["opportunity"]}
        except:
            print("failed at ", page["url"]["href"])

data = prepare(j)
cc_full = pd.json_normalize(data)
cc_full
```

```python
from w3lib.html import remove_tags
import locale
from locale import atof

lut = {40: 'Canada',
       16: 'Austria',
       1: 'United Kingdom',
       69: 'Estonia',
       75: 'France',
       60: 'Denmark',
       23: 'Belgium',
       42: 'Cayman Islands',
       82: 'Germany',
       84: 'Gibraltar',
       129: 'Luxembourg',
       156: 'Netherlands',
       178: 'Portugal',
       166: 'Norway',
       122: 'Latvia',
       231: 'United States of America',
       212: 'Switzerland',
       205: 'Spain',
       198: 'Singapore',
       211: 'Sweden',
       106: 'Ireland'}

clean = final_full.copy()
clean["country"] = clean.country.map(lambda x: lut[int(x)], na_action="ignore")
clean[money] = clean[money].map(lambda x: float(x[0])/100, na_action="ignore")
clean[mark_up] = clean[mark_up].map(remove_tags, na_action="ignore")
locale.setlocale(locale.LC_NUMERIC, 'en_GB')
clean[numeric] = clean[numeric].astype(str).map(atof, na_action="ignore")
dates = ["dates.activated", "dates.funded", "dates.closed"]
clean[dates] = clean[dates].map(pd.to_datetime)
clean[categorical] = clean[categorical].convert_dtypes("categorical")
clean[["eis", "seis"]] = clean[["eis", "seis"]].map(
    lambda x: x if x.__class__ != list else "_".join(x), na_action="ignore")
clean = clean.convert_dtypes()

dict(clean[core].iloc[0])
```

```python
import numpy as np
from functools import partial
from currency_converter import CurrencyConverter,ECB_URL
c = CurrencyConverter(fallback_on_missing_rate=True,fallback_on_missing_rate_method="linear_interpolation")
def f(x):
    try:
        return c.convert(amount=1,currency=x[0],date=x[1])
    except:
        return pd.NA
    
params = ["currency","dates.funded"]
d = clean[params+money].convert_dtypes().set_index(params)#.dropna()
d[d.add_suffix("_adj_EUR").columns] = d.mul(d.index.map(f,na_action="ignore"),axis=0).map(partial(np.round,decimals=2))
money2 = d.columns
t = clean.drop(columns=money).join(d[sorted(d.columns)].reset_index(drop=True))
a = t["Amount"].str.split("^(€|£)+",regex=True,expand=True).replace("",pd.NA).dropna(how="all",axis=1)
a.loc[:,1] = a.loc[:,1].map({"£":"GBP","€":"EUR"})
locale.setlocale(locale.LC_NUMERIC, 'en_GB')
a.loc[:,2] = a.loc[:,2].map(atof)
b = t[["Date","dates.funded"]].min(axis=1)
f = CurrencyConverter(ECB_URL,fallback_on_missing_rate=True,fallback_on_missing_rate_method="linear_interpolation")
t["Amount_adj_EUR"] = pd.concat((a,b),axis=1).apply(lambda x: f.convert(date=x[0],amount=x[2],currency=x[1]),axis=1).round(2)
first = ["id","name","url","Date",*dates,*sorted(money2)]
t = t[[*first,*sorted(set(t.columns)-set(first))]]
t.drop(columns=complex).to_csv("../data/adjusted/clean_full.csv", encoding="utf8",index=False)
t
```
