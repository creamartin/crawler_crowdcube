```python
import json
import pandas as pd
with open("./pitches.json") as f:
    j = json.load(f)


def prepare(arr):
    for page in arr:
        try:
            yield {"src_url": page["src_url"], "url": page["url"]["href"], **page["initialState"]["opportunity"]["opportunity"]}
        except:
            print("failed at ", page["url"]["href"])


data = prepare(j)
cc_full = pd.json_normalize(data)
```

```python
text = cc_full["the_people.content"].dropna().iloc[0]
```

```python
sections = [section.strip() for section in text.split("\n\n") if section.strip()]

# Initialize a list to store individual records
records = []

# Iterate through each section to extract information
for section in sections:
    lines = section.split('\n')
    name = lines[0].strip(',')
    job_title = lines[1].strip()
    experience = ' '.join(lines[2:]).strip()

    # Create a dictionary for the record
    record = {
        "Name": name,
        "Job Title": job_title,
        "Experience": experience
    }

    # Append the record to the list
    records.append(record)

# Convert the list of records to JSON
json_data = json.dumps(records, indent=2)

# Print the JSON data
print(json_data)
```
